```python
import numpy as np
```

```python
#Q1
arr = np.array([1, 2, 3, 6, 4, 5])
print(arr)
print(arr[::-1])

print()
```

```python
#Q2
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])
print(arr1)
print(arr2)
print(arr1 == arr2)

print()
```

```python
#Q3 (i)
x = np.array([1,2,3,4,5,1,2,1,1,1])
print(x)
values, counts = np.unique(x, return_counts=True)
print(values)
print(counts)
max_value = values[counts.argmax()]
print("Most frequent value:", max_value)
indices = np.where(x == max_value)
print("Indices:", indices[0])

print()
```

```python
#Q3(ii)
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])
print(y)
values, counts = np.unique(y, return_counts=True)
print(values)
print(counts)
max_value = values[counts.argmax()]
print("Most frequent value:", max_value)
indices = np.where(y == max_value)
print("Indices:", indices[0])

print()
```

```python
# Q4
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(gfg)
print("Sum of all elements:", gfg.sum())
print("Row wise sum:")
print(gfg.sum(axis=1))
print("Column wise sum:")
print(gfg.sum(axis=0))
```
